```python
import csv
import numpy as np
ls = []
with open('/home/manik/stock_analysis/Stock_extraction/GOOG_5Y.csv') as csvfile:
    rdr = csv.reader(csvfile)
    for row in rdr:
        ls.append(row)
data = np.array(ls)
data_open = data[1:,1:2]
```

```python
data_vals = data[1:, 1:].astype(float)
```

```python
op = data_vals[:, 0:1]
hi = data_vals[:, 1:2]
lo = data_vals[:, 2:3]
cl = data_vals[:, 3:4]
vo = data_vals[:, 4:5]
```

```python
frac_hi = (hi-op)/op
frac_lo = (op - lo)/op
frac_ch = (cl - op)/op
```

```python
d = 10
```

```python
print frac_hi.shape
print frac_lo.shape
print frac_ch.shape
```

```python
frac_hi = np.reshape(frac_hi,1257)
frac_lo = np.reshape(frac_lo,1257)
frac_ch = np.reshape(frac_ch,1257)
```

```python
X = np.column_stack([frac_ch, frac_hi, frac_lo])
```

```python
tr_set = X[0:500]
test_set = X[500:600]
```

```python
from hmmlearn import hmm
model = hmm.GaussianHMM(n_components=4, covariance_type="diag", n_iter=200)
model.fit(tr_set)
```

```python
print model.transmat_
```

```python
model.monitor_.converged
```

```python
possible_frac_ch = np.linspace(-0.1, 0.1, 50)
possible_frac_lo = np.linspace(0.0, 0.1, 10)
possible_frac_hi = np.linspace(0.0, 0.1, 10)
```

```python
def list_of_vals(testing_set, init, window):
    ls = []
    i = init
    while(i < init+window):
        ls.append(testing_set[i])
        i = i+1
    return ls
```
